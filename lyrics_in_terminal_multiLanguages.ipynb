```python
# import pandas
import pandas as pd
```

```python
# read it as a data frame
df_english = pd.read_fwf('/content/lyrics.txt', header=None)
```

# II.Translation

```python
# refer to https://pypi.org/project/youtube-transcript-api/
# install libraries
!pip install googletrans==3.1.0a0
```

```python
# convert a series to a list
english_list = df_english.iloc[:,0].tolist()
english_list
```

```python
# import a library
from googletrans import Translator
```

```python
# translate english to japanese
translator = Translator()
japanese_list = [translator.translate(item, dest='ja').text for item in english_list]
japanese_list
```

```python
# translate english to nepali
translator = Translator()
nepali_list = [translator.translate(item, dest='ne').text for item in english_list]
nepali_list
```

```python
# translate english to simplified chinese
translator = Translator()
chinese_list = [translator.translate(item, dest='zh-CN').text for item in english_list]
chinese_list
```

```python
# translate english to vietnamese
translator = Translator()
vietnamese_list = [translator.translate(item, dest='vi').text for item in english_list]
vietnamese_list
```

```python
# create a data frame from the lists
english_multi = pd.DataFrame(list(zip(english_list, japanese_list, nepali_list, chinese_list, vietnamese_list)),
                             columns = ['english', 'japanese', 'nepali', 'chinese', 'vietnamese'])
english_multi
```

```python
# Save it as a CSV file
english_multi.to_csv('lyrics_multi.csv')
```
