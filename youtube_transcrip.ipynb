```python
# import a library
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
# instantiate a TranscriptList object 
transcript_list = YouTubeTranscriptApi.list_transcripts('BR_dUDf8g9k')
transcript_list
```

```python
# fetch the english transcript
transcript = transcript_list.find_transcript(['en'])
transcript_fetched = transcript.fetch()
transcript_fetched
```

```python
# get english texts
transcript_text = [item['text'] for item in transcript_fetched]
transcript_text
```

```python
# translate english to japanese
transcript = transcript_list.find_transcript(['en'])
translated_transcript = transcript.translate('ja')
transcript_translated = translated_transcript.fetch()
```

```python
# get japanese transcripts using the youtube transcript api
japanese_youtube = [item['text'] for item in transcript_translated]
japanese_youtube
```

```python
# translate texts line by line
```

```python
# import libraries
from googletrans import Translator, constants
from pprint import pprint
import pandas as pd
```

```python
translator = Translator(service_urls=[
  'translate.google.com'])
translator = Translator()
```

```python
# translate text line by line
translations = translator.translate(transcript_text, dest='ja')
for translation in translations:
   print(translation.origin, ' -> ', translation.text)
```

```python
## create a data frame
# create a list of english texts
text_english = [item.origin for item in translations]
text_english
```

```python
# create a list of japanese texts
text_japanese = [item.text for item in translations]
text_japanese
```

```python
# create a data frame
df1 = pd.DataFrame(list(zip(text_english, text_japanese)),
                   columns = ['english', 'japanese'])
df1
```

```python
# add japanese translated by youtube as a new column
df2 = pd.DataFrame({'japanese_youtube': japanese_youtube})
df3 = pd.concat([df1, df2], axis=1) 
df3
```

```python
# save it as a CSV file
df3.to_csv('mytranscript.csv')
```
